should return the index of element to be searched in the list.


```python
def sequential_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Sorted list
numbers = [10, 20, 30, 40, 50, 60]

key = int(input("Enter element to search: "))

result = sequential_search(numbers, key)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

<h3>3) Write a function to calculate the factorial of a number (both iterative and recursive).


```python
# Iterative
def fact(n):
    f = 1
    for i in range(1, n + 1):
        f *= i
    return f

print(fact(5))

# Recursive
def fact_rec(n):
    if n == 1:
        return 1
    return n * fact_rec(n - 1)

print(fact_rec(5))
```

calculate area and perimeter.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r = Rectangle(10, 5)
print("Area:", r.area())
print("Perimeter:", r.perimeter())
```

<h3>5) Create a class Circle with method to calculate area. Overload the + operator to add two
circles (add their radii).

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def __add__(self, other):
        return Circle(self.radius + other.radius)

c1 = Circle(3)
c2 = Circle(4)

c3 = c1 + c2

print("Area of Circle 1:", c1.area())
print("Area of Circle 2:", c2.area())
print("Radius after adding circles:", c3.radius)
```

<h3>3) Create a class Book with special methods __str__, __len__, and __eq__.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return self.title

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.pages == other.pages

b1 = Book("Python", 200)
b2 = Book("Java", 200)

print(b1)
print(len(b1))
print(b1 == b2)
```

<h3>4) Create a class Temperature with property decorators for Celsius and Fahrenheit
conversion.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

t = Temperature(0)
print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)
```

<h3>1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle,
Circle, and Triangle that implement the area method.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Triangle(Shape):
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * self.b * self.h

r = Rectangle(10, 5)
c = Circle(3)
t = Triangle(4, 6)

print("Rectangle Area:", r.area())
print("Circle Area:", c.area())
print("Triangle Area:", t.area())
```

<h3>2) Create a class Library that manages a collection of books. Implement methods to add
books, remove books, search books, and display all books.

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)

    def search_book(self, book):
        if book in self.books:
            print("Book found")
        else:
            print("Book not found")

    def display_books(self):
        print("Books in library:")
        for b in self.books:
            print(b)

lib = Library()
lib.add_book("Python")
lib.add_book("Java")

lib.display_books()
lib.search_book("Python")
lib.remove_book("Java")
lib.display_books()
```
